### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import random
import json
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import NuSVC, LinearSVC
from sklearn.metrics import *
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.model_selection import learning_curve, validation_curve, cross_val_score, KFold, LeaveOneOut, ShuffleSplit, StratifiedKFold, GroupKFold, GridSearchCV
from sklearn import svm
from sklearn.feature_selection import SelectKBest, chi2, VarianceThreshold
from sklearn.calibration import CalibratedClassifierCV
```

```python
f = open('directories.json')
data = json.load(f)

TRAINDIR = data["TRAINDIR"]
TESTDIR = data["TESTDIR"]
f.close()

TWO_CATEGORIES = ["NORMAL", "PNEUMONIA"]
THREE_CATEGORIES = ["NORMAL", "VIRUS", "BACTERIA"]
```

```python
def create_training_data(url, categories, imgSize, nbImage=None):
    data = []
    for category in categories:
        path = os.path.join(url, category)
        shuffled_dir = os.listdir(path)
        random.shuffle(shuffled_dir)
        class_num = categories.index(category)
        imgForLoop = range(0, nbImage) if nbImage else shuffled_dir
        test = "avec bnImage" if nbImage else "sans nbImage"
        for img in imgForLoop:
            try:
                img_array = cv2.imread(os.path.join(path, shuffled_dir[img]), cv2.IMREAD_GRAYSCALE) if nbImage else cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array, (imgSize, imgSize))
                data.append([new_array, class_num])
            except Exception as e:
                pass
    return process(data, imgSize)

def process(data, imgSize):
    X = []
    y = []
    for features, label in data:
        X.append(features)
        y.append(label)
    X = np.array(X).reshape(-1, imgSize*imgSize)
    y = np.array(y)
    X.shape
    return [X, y]

def showStatistics(models, labels, X_test, y_test):
    predictions = []
    for index, model in enumerate(models):
        pred = model.predict(X_test)
        predictions.append([labels[index], 
                            precision_score(y_test, pred, average='weighted'), 
                            recall_score(y_test, pred, average='weighted'), 
                            f1_score(y_test, pred, average='weighted')])
        df = pd.DataFrame(data=predictions, columns=["Models", "Precision", "Recall", "F1"])
    df.set_index("Models", inplace = True)   
    return df
```

### Choose image sizes

```python
resultatImg = []
label= []
for i in range (50, 350, 50):
    X, y = create_training_data(TRAINDIR, TWO_CATEGORIES, i)
    resultatImg.append(cross_val_score(KNeighborsClassifier(), X, y, cv=4).mean())
    label.append(i)
graphImg = np.array(resultatImg)
df = pd.DataFrame(data=graphImg, columns=['value'], index=label)
plt.figure(figsize=(15, 8))
s = sns.lineplot(data=df)
s.set_xlabel("pixel size")
s.set_ylabel("cross val score")
s.set_title("impact of image size")
```

We can see that the model has better performance with 200 pixels. We will now use this size. But for better processing time we are using 100 pixels.

```python
IMG_SIZE = 100
```

### Choose number of images for dataset

```python
X_train, y_train = create_training_data(TRAINDIR, TWO_CATEGORIES, IMG_SIZE)
train_sizes, train_scores, test_scores = learning_curve(estimator=KNeighborsClassifier(), X=X_train, y=y_train,
                                                       cv=10, train_sizes=np.linspace(0.1, 1.0, 10),
                                                     n_jobs=1)
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)
plt.figure(figsize=(15, 8))
plt.plot(train_sizes, train_mean, color='blue', marker='o', markersize=5, label='Training Accuracy')
plt.fill_between(train_sizes, train_mean + train_std, train_mean - train_std, alpha=0.15, color='blue')
plt.plot(train_sizes, test_mean, color='green', marker='+', markersize=5, linestyle='--', label='Validation Accuracy')
plt.fill_between(train_sizes, test_mean + test_std, test_mean - test_std, alpha=0.15, color='green')
plt.title('Learning Curve')
plt.xlabel('Training Data Size')
plt.ylabel('Model accuracy')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

As the score doesn't increase much when there are more than 1300 images in each categories of dataset, we will now import only 1300 images per categories. 

```python
IMG_PER_CATEGORY = 1300
```

```python
X_train, y_train = create_training_data(TRAINDIR, TWO_CATEGORIES, IMG_SIZE, IMG_PER_CATEGORY)
X_test, y_test = create_training_data(TESTDIR, TWO_CATEGORIES, IMG_SIZE, IMG_PER_CATEGORY)
print(f'Shape of X_train: {X_train.shape}')
print(f'Shape of X_test: {X_test.shape}')
```

### Check normalization utility

```python
not_normalized = KNeighborsClassifier().fit(X_train, y_train)
normalized = make_pipeline(
    MinMaxScaler(),
    KNeighborsClassifier()).fit(X_train, y_train)
```

```python
normalization_comparision = showStatistics([not_normalized, normalized], 
                                               ["Without normalization", "With normalization"], 
                                               X_test, y_test)
normalization_comparision
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(data=normalization_comparision)
```

We observe a slight improvement with normalization. We will now use it.

### Check Feature selection utility

```python
model_without_FS = make_pipeline(
    MinMaxScaler(), 
    KNeighborsClassifier(n_neighbors=4, metric='manhattan')).fit(X_train, y_train)

model_with_VarianceThreshold = make_pipeline(
    VarianceThreshold(threshold=1500),
    MinMaxScaler(), 
    KNeighborsClassifier(n_neighbors=4, metric='manhattan')).fit(X_train, y_train)

model_with_SelectKBest = make_pipeline(
      SelectKBest(chi2, k=100),
     KNeighborsClassifier(n_neighbors=4, metric='manhattan')).fit(X_train, y_train)
```

```python
feature_selection_comparision = showStatistics([model_without_FS, model_with_VarianceThreshold, model_with_SelectKBest], 
                                               ["Simple KNN", "KNN with VarianceThreshold", "KNN with SelectKBest"], 
                                               X_test, y_test)
feature_selection_comparision
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(data=feature_selection_comparision)
```

In our case, feature selection seems to descrease the accuracy of our model. We will not use it. 

### Compare models

```python
knn = make_pipeline(
    MinMaxScaler(),
    KNeighborsClassifier(n_neighbors=4, metric='manhattan'))
knn.fit(X_train, y_train)
print(f'KNeighborsClassifier accuracy: {knn.score(X_test, y_test)}')
```

```python
clfnS = make_pipeline(
MinMaxScaler(),
NuSVC(nu=0.5, probability=True))
clfnS.fit(X_train, y_train)
print(f'NuSVC accuracy: {clfnS.score(X_test, y_test)}')
```

```python
clfnL = make_pipeline(
StandardScaler(),
CalibratedClassifierCV(LinearSVC()))
clfnL.fit(X_train, y_train)
print(f'LinearSVC accuracy: {clfnL.score(X_test, y_test)}')
```

```python
clf = make_pipeline(
StandardScaler(),
HistGradientBoostingClassifier(max_iter=100))
clf.fit(X_train, y_train)
print(f'HistGradientBoostingClassifier accuracy: {clf.score(X_test, y_test)}')
```

```python
model_comparison = showStatistics([knn, clfnS, clfnL, clf], ["knn", "clfnS", "clfnL", "clf"], X_test, y_test)
model_comparison
```
